```python
# Dependencies and Setup
import pandas as pd
```

```python
# Files to Load 
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load,index_col="school_name")
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete=school_data_complete.sort_values(by="school_name")
```

```python
#Get all school names 
school_names=school_data_complete["school_name"].unique()
```

```python
#Calculate the number of schools   
total_number_of_schools=len(school_names)
print(total_number_of_schools)
```

```python
#Calculate the number of students 
total_number_of_students=school_data_complete["student_name"].count()
print(total_number_of_students)
```

```python
#Calculate total budget
filter_school_df=school_data_complete.drop_duplicates(subset='school_name')
total_budget=filter_school_df["budget"].sum()
print(total_budget)
```

```python
#Calculate average math score 
average_math_score=school_data_complete["math_score"].mean()
print(average_math_score)
```

```python
#Calculate average reading score 
average_reading_score=school_data_complete["reading_score"].mean() 
print(average_reading_score)
```
