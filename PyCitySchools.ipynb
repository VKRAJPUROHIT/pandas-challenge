```python
# Dependencies and Setup
import pandas as pd
```

```python
# Files to Load 
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load,index_col="school_name")
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_data_complete=school_data_complete.sort_values(by="school_name")
school_data_complete.head()
```

```python
#Check to see if there are any missing values by displaying True or False
school_data_complete.isna()
```

```python
#Get all school names 
school_names=school_data_complete["school_name"].unique()
```

```python
#Calculate the number of schools   
total_number_of_schools=len(school_names)
print(total_number_of_schools)
```

```python
#Calculate the number of students 
total_number_of_students=school_data_complete["student_name"].count()
print(total_number_of_students)
```

```python
#Calculate total budget
filter_school_df=school_data_complete.drop_duplicates(subset='school_name')
total_budget=filter_school_df["budget"].sum()
print(total_budget)
```

```python
#Get statistics of the data frame 
school_data_complete.describe()
```

```python
#Calculate average math score 
average_math_score=school_data_complete["math_score"].mean()
print(average_math_score)
```

```python
#Calculate average reading score 
average_reading_score=school_data_complete["reading_score"].mean() 
print(average_reading_score)
```

```python
#Calculate number of students with passing math score 
names_math=school_data_complete[(school_data_complete['math_score']>=70)]
number_of_students_with_passing_math_score=len(names_math)
print(number_of_students_with_passing_math_score)
```

```python
#Calculate percentage of students with passing math score
percent_math_70_or_greater=number_of_students_with_passing_math_score/total_number_of_students*100
print(percent_math_70_or_greater)
```

```python
#Calculate number of students with passing reading score 
names_reading=school_data_complete[(school_data_complete['reading_score']>=70)]
number_of_students_with_passing_reading_score=len(names_reading) 
print(number_of_students_with_passing_reading_score)
```

```python
#Calculate percentage of students with passing reading score
percent_reading_70_or_greater=number_of_students_with_passing_reading_score/total_number_of_students*100 
print(percent_reading_70_or_greater)
```

```python
#Get the names of students with passing math AND reading
names_math_and_reading=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]
names_math_and_reading.head()
```

```python
#Calculate the number of students with passing math AND reading
number_math_and_reading=len(names_math_and_reading)
print(number_math_and_reading)
```

```python
#Calculate the percentage of students passing math AND reading  
percent_math_and_reading=number_math_and_reading/total_number_of_students*100
print(percent_math_and_reading)
```
